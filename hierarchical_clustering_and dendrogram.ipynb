```python
from matplotlib import pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
import numpy as np
%matplotlib inline
np.set_printoptions(precision=5, suppress=True)
```

```python
np.random.seed(4711)
a_mean = [10,0]
b_mean = [0,20]
cov = [[3,1],[1,4]]
a = np.random.multivariate_normal(a_mean, cov, size=[100,])
b = np.random.multivariate_normal(b_mean, cov, size=[50,])
```

```python
X = np.concatenate((a,b),)
```

```python
plt.scatter(X[:,0], X[:,1])
```

```python
Z = linkage(X, 'ward')
```

```python
from scipy.cluster.hierarchy import cophenet
from scipy.spatial.distance import pdist
```

```python
cophenet(Z, pdist(X))
```

```python
plt.figure(figsize=(25,10))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('sample index')
plt.ylabel('distance')
dendrogram(Z, leaf_rotation=90, leaf_font_size=8)
plt.show()
```

```python
plt.figure(figsize=(25,15))
plt.title('Hierarchical Clustering Dendrogram (Truncated)')
plt.xlabel('sample index')
plt.ylabel('distance')
dendrogram(Z, truncate_mode='lastp', p=12, orientation='top', leaf_rotation=0, show_leaf_counts=True, leaf_font_size=20, show_contracted=True)
plt.show()
```
